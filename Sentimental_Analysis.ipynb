# **Libraries :**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import sklearn
import re
```

```python
data=pd.read_csv('movie.csv')
```

# **Preprocessing :**

```python
X=data['text']
y=data['label']
```

## **Text cleaning :**

We are gonna begin with removing the non alphabetic characters, converting the text to lowercase and removing the stopwords.

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
import string
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
stop_words = stopwords.words('english')
punctuations = string.punctuation
```

```python
import re
nltk.download('wordnet')
```

```python
clean_data = []
for i in range(len(X)):
  text = re.sub('[^a-zA-Z]', ' ',X.iloc[i])
  text = text.lower().split()
  text = [lemmatizer.lemmatize(word) for word in text if (word not in stop_words) and (word not in punctuations)]
  text = ' '.join(text)
  clean_data.append(text)
```

## **Vectorizing :**

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
X_vec = count_vectorizer.fit_transform(clean_data)
```

# **Classification Model :**

We'll be using MultinomialNB, since it works good with text classification tasks.

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
model = MultinomialNB()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_vec,y, test_size = 0.3)
```

```python
model.fit(X_train,Y_train)
```

# **Evaluation :**

```python
y_pred = model.predict(X_test)

from sklearn.metrics import classification_report

classification = classification_report(Y_test,y_pred)
print(classification)
```
